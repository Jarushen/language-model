```python
import csv
import random
from faker import Faker
from datetime import datetime, timedelta

# Create a Faker instance with default locale (usually English)
fake = Faker()

def generate_transaction():
    # List of South African companies
    companies = [
        'Checkers', 'Woolworths', 'Pick n Pay', 'Spar', 'Shoprite',
        'Game', 'Makro', 'Builders Warehouse', 'Clicks', 'Dis-Chem',
        'Mr Price', 'Edgars', 'Jet', 'Truworths', 'Ackermans',
        'Boxer Superstores', 'Exclusive Books', 'HiFi Corp', 'Incredible Connection', 'Takealot'
    ]
    
    # Extensive list of locations in South Africa
    locations = [
        'Cape Town', 'Johannesburg', 'Durban', 'Pretoria', 'Port Elizabeth',
        'Soweto', 'East London', 'Polokwane', 'Pietermaritzburg', 'Bloemfontein',
        'Nelspruit', 'Kimberley', 'Mafikeng', 'Rustenburg', 'George',
        'Knysna', 'Paarl', 'Stellenbosch', 'Franschhoek', 'Hermanus',
        'Grahamstown', 'Uitenhage', 'Oudtshoorn', 'Claremont', 'Sea Point',
        'Hout Bay', 'Simonâ€™s Town', 'Kalk Bay', 'Muizenberg', 'Sunninghill'
    ]

    # Randomly choose a company, possibly removing internal spaces
    company = random.choice(companies)
    if random.choice([True, False]) and ' ' in company:
        company = company.replace(' ', '')

    # Randomly apply a prefix
    prefix = random.choice(['c*', 'b*', ''])

    # Choose a location
    location = random.choice(locations)

    # Optionally add a full card number
    include_card = random.choice([True, False])
    if include_card:
        card_number = f'{random.randint(1000, 9999)} {random.randint(1000, 9999)} {random.randint(1000, 9999)} {random.randint(1000, 9999)}'
        stmt_text = f"{prefix}{company} {location} {card_number}"
    else:
        stmt_text = f"{prefix}{company} {location}"

    return stmt_text

def generate_transaction_date():
    # Generate a random date within the last year
    start_date = datetime.now() - timedelta(days=365)
    random_date = start_date + timedelta(days=random.randint(0, 365))
    return random_date.strftime('%Y-%m-%d')

def generate_transaction_amount():
    # Generate a realistic transaction amount
    return round(random.uniform(5.0, 500.0), 2)

def generate_transactions(num_transactions):
    transactions = []
    for _ in range(num_transactions):
        stmt_tx = generate_transaction()
        txn_date = generate_transaction_date()
        txn_amount = generate_transaction_amount()
        transactions.append([stmt_tx, txn_date, txn_amount])
    return transactions

def save_transactions_to_csv(transactions, filename='transactions.csv'):
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['STMT_TX', 'TRANSACTION_DATE', 'TRANSACTION_AMOUNT'])
        writer.writerows(transactions)

# Example usage:
num_transactions = 1000
transactions = generate_transactions(num_transactions)
save_transactions_to_csv(transactions)
```
